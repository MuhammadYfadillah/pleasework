### Cleaning Data

#### Pada tahap ini saya melakukan pembersihan data. dikarenakan yang menjadi fokus analisis hanya pada PM2.5 dan PM10, maka kolom-kolom seperti No, hour, SO2, NO2, CO, O3, TEMP, DEWP, RAIN, WD, WSPM, dan PRES bisa dihapus. Metode Imputasi akan digunakan untuk mengisi missing value.

```python
shunyi_df.drop(['No','hour', 'day', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'DEWP', 'RAIN', 'wd', 'WSPM', 'PRES'], axis=1, inplace=True) #menghapus kolom yang tidak diperlukan
```

```python
tiantan_df.drop(['No','hour', 'day', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'DEWP', 'RAIN', 'wd', 'WSPM', 'PRES'], axis=1, inplace=True) #menghapus kolom yang tidak diperlukan
```

```python
shunyi_df['PM2.5']  = shunyi_df['PM2.5'].fillna(shunyi_df['PM2.5'].mean()) #mengisi missing value pada kolom PM2.5
shunyi_df['PM10']  = shunyi_df['PM10'].fillna(shunyi_df['PM10'].mean()) #mengisi missing value pada kolom PM10
tiantan_df['PM2.5']  = tiantan_df['PM2.5'].fillna(tiantan_df['PM2.5'].mean()) #mengisi missing value pada kolom PM2.
tiantan_df['PM10']  = tiantan_df['PM10'].fillna(tiantan_df['PM10'].mean()) #mengisi missing value pada kolom PM10
```
